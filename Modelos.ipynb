## Read data

```python
df = pd.read_csv('datapreprocess.csv', index_col=[0])
df.head()
```

### Data Preprocessing

```python
#Seleccionamos las variables categoricas
cat = df.select_dtypes('O')

ohe = OneHotEncoder(sparse_output=False)

#entrenamiento y transformacion
cat_ohe = ohe.fit_transform(cat)

#Colocar el nombre a cada columa
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features=cat.columns)).reset_index(drop = True)
cat_ohe.head()
```

```python
#Numeric cols
num = df.select_dtypes('number').reset_index(drop = True)
num.head(1)
```

```python
df_ml = pd.concat([cat_ohe, num], axis =1)
df_ml.head()
```

```python
df_ml.to_csv('dataforml.csv')
```

### Split dataset
1. X and Y
1. Train and testing set

```python
X = df_ml.drop('abandono', axis = 1)
y = df_ml.abandono
y.value_counts()
```

```python
#2. Train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import accuracy_score, recall_score, precision_score
```

```python
#Enternamiento y evaluación de los modelos
models = [
    ("Logistic Regression", LogisticRegression()),
    ("Stochastic Gradient Descent", SGDClassifier()),
    ("Decision Tree Classifier", DecisionTreeClassifier()),
    ("Random Forest Classifier", RandomForestClassifier()),
    ("Gradient Boosting Classifier", GradientBoostingClassifier()),
    ("AdaBoost Classifier", AdaBoostClassifier()),
    ("Extra Trees Classifier", ExtraTreesClassifier()),
    ("SVM", SVC()),
    ("KNeighbors Classifier", KNeighborsClassifier()),
    ("Gaussian Naive Bayes", GaussianNB()),
    ("MLP Classifier", MLPClassifier()),
    ("XGBoost Classifier", XGBClassifier()),
    ("LightGBM Classifier", LGBMClassifier()),
    ("CatBoost Classifier", CatBoostClassifier(silent=True)),
    ("Quadratic Discriminant Analysis", QuadraticDiscriminantAnalysis())
]

results = []
for name, model in models:
    pipeline = Pipeline([
        ("scaler", StandardScaler()),
        ("model", model)
    ])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    results.append((name, score, recall, precision))

# Print results with both accuracy and recall
for name, accuracy, recall, precision in results:
    print(f"{name}: Accuracy - {accuracy:.4f}, Recall - {recall:.4f}, precision - {precision:.4f}")
```

```python
#Print Result , Create a DataFrame
df_results = pd.DataFrame(results, columns = ['modelo', 'accuracy', 'recall', 'precision'])
df_results.sort_values(by = 'accuracy', ascending = False)
```

#### Hemos Evaluado el:

**Accuracy:** representa el porcentaje total de valores correctamente clasificados, tanto positivos como negativos  

**Recall:** mide la proporción de los casos positivos reales que fueron correctamente identificados por el modelo    

**Precisión:** La métrica de precisión es utilizada para poder saber qué porcentaje de valores que se han clasificado como positivos son realmente positivos

## Técnica de Balancemiento de datos
### Peso en las clases
**Ajuste de Parámetros del modelo:** Consiste en ajustar parametros ó metricas del propio algoritmo para intentar equilibrar a la clase minoritaria penalizando a la clase mayoritaria durante el entrenamiento. Ejemplos on ajuste de peso en árboles, también en logisticregression tenemos el parámetro class_weight= “balanced” que utilizaremos en este ejemplo. No todos los algoritmos tienen estas posibilidades. En redes neuronales por ejemplo podríamos ajustar la métrica de Loss para que penalice a las clases mayoritarias.

```python
#Enternamiento y evaluación de los modelos
models = [
    ("Logistic Regression", LogisticRegression(class_weight='balanced')),
    ("Stochastic Gradient Descent", SGDClassifier(class_weight='balanced')),
    ("Decision Tree Classifier", DecisionTreeClassifier(class_weight='balanced')),
    ("Random Forest Classifier", RandomForestClassifier(class_weight='balanced')),
    #("Gradient Boosting Classifier", GradientBoostingClassifier()),
    #("AdaBoost Classifier", AdaBoostClassifier()),
    ("Extra Trees Classifier", ExtraTreesClassifier(class_weight='balanced')),
    #("SVM", SVC(class_weight='balanced')),
    #("KNeighbors Classifier", KNeighborsClassifier()),
    #("Gaussian Naive Bayes", GaussianNB()),
    #("MLP Classifier", MLPClassifier()),
    #("XGBoost Classifier", XGBClassifier()),
    ("LightGBM Classifier", LGBMClassifier(class_weight='balanced'))
    #("CatBoost Classifier", CatBoostClassifier(silent=True, class_weights='balanced'))
    #("Quadratic Discriminant Analysis", QuadraticDiscriminantAnalysis())
]

results = []
for name, model in models:
    pipeline = Pipeline([
        ("scaler", StandardScaler()),
        ("model", model)
    ])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    results.append((name, score, recall, precision))

# Print results with both accuracy and recall
for name, accuracy, recall, precision in results:
    print(f"{name}: Accuracy - {accuracy:.4f}, Recall - {recall:.4f}, precision - {precision:.4f}")
```

```python
#Print Result , Create a DataFrame
df_p = pd.DataFrame(results, columns = ['modelo', 'accuracy', 'recall', 'precision'])
df_p.sort_values(by = 'accuracy', ascending = False)
```

**Observaciones**:    
- Para este caso los modelos que se entrenaron con el parametro de penalización de las clases mejoraron el recall, pero disminuyeron el accuracy

## Estrategia: Oversampling de la clase minoritaria
1. Random Over-Sampling (ROS):  
  - Ventaja: Simple y rápido de implementar.
  - Desventaja: Puede generar sobreajuste.
2. SMOTE (Synthetic Minority Over-sampling Technique):  
   - Ventaja: Más sofisticado que ROS, con potencial para una mejor generalización.
  - Desventaja: Puede introducir puntos irreales en la distribución de la clase minoritaria.  
3. ADASYN (Adaptive Synthetic Minority Over-sampling Technique):  
  - Ventaja: Reduce el sobreajuste y mejora la representatividad de la clase minoritaria.
  - Desventaja: Más complejo y computacionalmente costoso que ROS y SMOTE.
4. Borderline-SMOTE (BL-SMOTE):
  - Ventaja: Mejora la clasificación en casos difíciles.
  - Desventaja: Similar a ADASYN en complejidad y costo computacional


**Para este caso utilizare el ADSYN y el BL-SMOTE**
